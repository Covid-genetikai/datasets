```python
import json
import pandas as pd
```

```python
df = pd.read_csv("ncbi_sgene_good_unique.csv")
```

```python
df[df['accession'] == '2246_6593532f926e48cc68421ef20a33018c']["accession"].values[0]
```

```python
bin_tree = json.load(open("tree.json"))
```

```python
bin_tree
```

```python
# rename_map = {}
```

```python
# Convert json tree into more friendly format
leaves = {} # gvid : accession
for o in bin_tree["objects"]:
    gvid = o["_gvid"]
    name = o["name"]#.replace("_", ".")
#     print(name)
    accession = df[df['accession'] == name]["accession"].values[0]
    leaves[gvid] = accession
    
#     rename_map[o["name"]]=accession    
```

```python
# [(parent, child), ...]
pairs = []
for edge in bin_tree["edges"]:
    pairs.append({
        "parent": leaves[edge["tail"]],
        "child": leaves[edge["head"]]
    })    
```

```python
df_pairs = pd.DataFrame.from_dict(pairs)
```

```python
df_pairs
```

```python
df_pairs.to_csv("pairs.csv", index=False)
```
